# Importando as bibliotecas

```python
import pyautogui
import time
import pandas as pd
import pyperclip
```

# Automatizando a obtenção da base de dados

```python
link = 'https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema'
pyautogui.PAUSE = 1

# Abrir o sistema
pyautogui.hotkey('ctrl', 't')
pyautogui.write(link)
pyautogui.press('enter')
time.sleep(5)

# Preencher login
pyautogui.click(x=657, y=378)
pyautogui.write('meu_login')

# Preencher a senha
pyautogui.click(x=644, y=447)
pyautogui.write('minha_senha')

# Acessar
pyautogui.click(x=692, y=519)
time.sleep(10)

# Selecionar arquivo
pyautogui.click(x=477, y=302)

# Realizar o download da base de dados
pyautogui.click(x=1165, y=190)
pyautogui.click(x=946, y=592)

time.sleep(5)
```

# Importando a base de dados

```python
tabela = pd.read_csv('C://Users/milan/Downloads/Compras.csv', sep=';')
display(tabela)
total_gasto = tabela['ValorFinal'].sum()
quantidade = tabela['Quantidade'].sum()
preco_medio = total_gasto / quantidade
```

# Automatizando o envio do email

```python
gmail = 'https://mail.google.com/mail/u/0/'
email = 'milaniguilherme@hotmail.com'
assunto = 'Relatório de compras'
mensagem = f"""
Prezados,
Segue resumo das compras:

Gasto total: R${total_gasto:,.2f}
Quantidade: {quantidade}
Preço médio: R${preco_medio:,.2f}

Att,
Guilherme Milani Silva
"""
pyautogui.PAUSE = 1

# Entrando no gmail
pyautogui.hotkey('ctrl', 't')
pyautogui.write(gmail)
pyautogui.press('enter')
time.sleep(5)

# Clicando em escrever
pyautogui.click(x=159, y=204)
time.sleep(2)

# Escrevendo o e-mail do destinátario
pyautogui.write(email)
pyautogui.press('enter')
pyautogui.press('tab')
time.sleep(2)

# Escrevendo o assunto do e-mail
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')
time.sleep(2)

# Escrevendo a mensagem do e-mail
pyperclip.copy(mensagem)
pyautogui.hotkey('ctrl', 'v')

# Clicando em enviar
pyautogui.click(x=819, y=689)
```
