# Importando bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importando base de dados

```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

# Traduzindo as colunas e os idiomas

```python
print(tmdb.columns)
tmdb.original_language.unique()
```

```python
traducao = {'budget': 'Orçamento', 'genres': 'Gênero', 'homepage': 'Site', 'id': 'Identificador', 'keywords': 'Palavras chaves',
            'original_language': 'Idioma', 'original_title': 'Titulo', 'overview': 'Sinopse',
            'popularity': 'Popularidade', 'production_companies': 'Produtora', 'production_countries': 'País de produção',
            'release_date': 'Data de lançamento', 'revenue': 'Receita', 'runtime': 'Tempo de duração', 
            'spoken_languages': 'Dublagens', 'status': 'Status', 'tagline': 'Slogan', 'title': 'Título',
            'vote_average': 'Nota', 'vote_count': 'Contagem'}
idiomas = {'en': 'Inglês', 'ja': 'Japonês', 'fr': 'Francês', 'zh': 'Mandarin', 'es': 'Espanhol', 'de': 'Alemão', 'hi': 'Hindu',
           'ru': 'Russo', 'ko': 'Creano', 'te': 'Polinésio', 'cn': 'Chinês', 'it': 'Italiano', 'nl': 'Holandês', 'ta': 'Tâmil',
           'sv': 'Suéco', 'th': 'Tailanês', 'da': 'Dinamarquês', 'xx': 'Xhosa', 'hu': 'Húngaro', 'cs': 'Tcheco',
           'pt': 'Português', 'is':' Islandês', 'tr': 'Turco', 'nb': 'Neonorueguês', 'af': 'Africano', 'pl': 'Polonês',
           'he': 'Hebreu', 'ar': 'Árabe', 'vi': 'Vietnamita', 'ky': 'Quirguiz', 'id': 'Indonésio', 'ro': 'Romeno',
           'fa': 'Persa', 'no': 'Norueguês', 'sl': 'Esloveno', 'ps': 'Pashto', 'el': 'Grego'}
```

```python
movies = tmdb.rename(columns = traducao)
movies['Idioma'] = movies['Idioma'].map(idiomas)
movies.head()
```

# Estudo dos idiomas originais

```python
movies['Idioma'].unique()
```

```python
m = movies['Idioma'].value_counts()
print(f'O idioma original mais frequente é o {m.idxmax()} com {max(m)} filmes. Seguido do {m.index[1]} com {m[1]} filmes!')
```

Podemos observar, que o inglês é o idioma original de filmes mais frequente no dataset, com mais de 4500 filmes produzidos. Esse era um resultado esperado, visto que os Estados Unidos, são referência nesse quesito. O francês vem logo em seguida, com 70 filmes. Levando em consideração, que a França é precursora no mundo do cinema, com a criação do próprio, pelos irmãos Lumière, é compreensível a quantidade de filmes em francês.

# Análises gráficas

```python
total_geral = m.sum()
total_ingles = m.loc['Inglês']
total_outros = total_geral - total_ingles
```

```python
dados = {'Língua': ['Inglês', 'Outras'],
        'Total': [total_ingles, total_outros]}
dados = pd.DataFrame(dados)
dados
```

## Gráfico de barras

```python
ax = sns.barplot(data=dados, x='Língua', y='Total')
ax.set_title('Inglês X Outros')
ax = ax
```

## Gráfico de Pizza

```python
ax = plt.pie(x=dados['Total'], labels=dados['Língua'])
plt.title('Inglês X Outras')
ax = ax
```

Com ambos os gráficos, é possivel observar a superioridade massiva do idioma Inglês, em relação ao restante.

# Análise de outros idiomas

```python
filmes_sem_lingua_original_em_ingles = movies.query("Idioma != 'Inglês'")
total_por_lingua_de_outros_filmes = movies.query("Idioma != 'Inglês'").Idioma.value_counts()
```

```python
ax = sns.catplot(x= 'Idioma', data = filmes_sem_lingua_original_em_ingles, aspect=2, kind='count',
            order=total_por_lingua_de_outros_filmes.index, palette="GnBu_d")
ax.set_xticklabels(labels=total_por_lingua_de_outros_filmes.index.drop_duplicates(), rotation=75)
ax.set_ylabels('Quantidade de filmes')
plt.title('Filmes em idiomas diferentes do inglês')
ax=ax
```

Com esse gráfico, é possivel observar a superioridade do idioma francês, em relação as outras línguas, excluindo o inglês, tendo mais que o dobro que o segundo colocado, o espanhol. O idioma português aparece como o decimo primeiro idioma com mais filmes nesse banco de dados, mostrando que a industria cinematográfica lusofônica não é tão relevante no cenário mundial, apesar de alguns filmes terem concorrido ao Oscar.

# Análise de outra base de dados, relacionada a filmes

```python
filmes = pd.read_csv('movies.csv')
filmes.columns = ['Id', 'Título', 'Gênero']
filmes.head()
```

```python
identifica = filmes.Id
film = pd.Index(filmes['Título'])
film = film.to_series(index=identifica)
```

```python
id_filmes = film.to_dict()
```

```python
notas = pd.read_csv('ratings.csv')
notas.columns = ['Usuário', 'Filme', 'Nota', 'Momento']
notas.Filme = notas.Filme.map(id_filmes)
notas.head()
```

# Distribuição de frequência

```python
ax = notas.Nota.plot(kind='hist', title='Distribuição de frequência de notas', figsize=(10, 7))
ax.set_xlabel('Notas')
ax.set_ylabel('Frequência')
ax=ax
```

```python
print(f'A média da notas é {notas.Nota.mean():.2f}')
print(f'A mediana da notas é {notas.Nota.median():.2f}')
print(f'A moda da notas é {notas.Nota.mode()[0]}')
```

O gráfico apresenta um comportamento assimétrico a esquerda. A moda é maior que a média e a mediana.

## Analisando as notas do usuário 1

```python
media = notas.query('Usuário == 1').Nota.mean()
```

```python
print(f'A média dde notas dado pelo usuário 1, é {media:.2f}')
```

## Média de notas por filme

```python
media_por_filme = notas.groupby('Filme').mean().Nota
media_por_filme.head()
```

## Gráfico de frequência

```python
ax = media_por_filme.plot(kind='hist', title='Média de notas por filme', figsize=(10, 7))
ax.set_xlabel('Notas')
ax.set_ylabel('Frequência')
ax=ax
```

```python
print(f'A nota média entre os filmes é {media_por_filme.mean():.2f}')
print(f'A mediana entre os filmes é {media_por_filme.median():.2f}')
print(f'A moda entre os filmes é {media_por_filme.mode()[0]}')
```

O gráfico apresenta um comportamento assimétrico a esquerda. A moda é maior que a mediana, que é maior que a média.

## Gráfico boxplot

```python
ax = sns.boxplot(data=media_por_filme, orient='h')
plt.title('Boxplot - Média por filme')
plt.figure(figsize=(10, 7))
ax=ax
```

O gráfico boxplot, reforça a informação fornecida pelo gráfico anterior. Assimetria a esquerda. A maior parte dos filmes tem nota acima de 3, o que indica uma boa qualidade.  

## Gráfico histograma

```python
ax = sns.displot(media_por_filme, kind='kde', aspect=2)
plt.title('Histograma - Média por filme')
ax=ax
```

# Comparando os filmes Toy Story e Jumanji

```python
filmes.head(2)
```

```python
notas_toy_story = notas.query('Usuário == 1')
notas_jumanji = notas.query('Usuário == 2')
```

```python
print(f'O filme Toy Story possui {len(notas_toy_story)}.')
print(f'O filme Jumanji possui {len(notas_jumanji)}.')
```

```python
print(f'A nota média do filme Toy Story é {notas_toy_story.Nota.mean():.2f}.')
print(f'A nota média do filme Jumanji é {notas_jumanji.Nota.mean():.2f}.')
```

```python
print(f'A mediana do filme Toy Story é {notas_toy_story.Nota.median()}.')
print(f'A mediana do filme Jumanji é {notas_jumanji.Nota.median()}.')
```

```python
print(f'A moda do filme Toy Story é {notas_toy_story.Nota.mode()[0]}.')
print(f'A moda do filme Jumanji é {notas_jumanji.Nota.mode()[0]}.')
```

```python
ax = sns.displot(notas_toy_story.Nota, aspect=2, kind='kde')
ax.set_ylabels('Frequência')
plt.title('Distribuição de frequência de notas - Toy Story')
ax=ax
```

```python
ax = sns.displot(notas_jumanji.Nota, kind='kde', aspect=2)
ax.set_ylabels('Frequência')
plt.title('Distribuição de frequência de notas - Jumanji')
ax=ax
```

É possivel observar que, o gráfico referente ao Toy Story, possui assimetria a esquerda. Fica evidente que a nota máxima 5, foi a mais frequente, o que indica a boa qualidade do filme. Em relação ao filme Jumanji, a curva se comporta aproximadamente como uma distribuição normal, com uma leve assimetria a esquerda. A nota 4 foi a mais frequente, o que confere uma boa qualidade ao filme, porém menor em relação ao Toy Story.

## Gráfico boxplot para os filmes, Toy Story e Jumanji

```python
ax = sns.boxplot(x='Usuário', y='Nota', data = notas.query('Usuário in (1,2)'))
plt.title('Boxplot (Toy Story X Jumanji)')
ax=ax
```

```python
print(f'Desvio padrão - Toy Story: {notas_toy_story.Nota.std():.2f}')
print(f'Desvio padrão - Jumanji: {notas_jumanji.Nota.std():.2f}')
```

Com esse gráfico, fica evidente a superioridade do filme Toy Story, em relação ao filme Jumanji. Basicamente, 75 % das pessoas que assistiram o filme, avaliaram com nota 4, enquanto apenas 50%, deram a mesma nota, para o filme Jumanji. É possivel perceber a existência de Outliers, para Toy Story, e não para o filme Jumanji. Esses outliers, podem ter sido gerados por um mecanismo distinto, ou até mesmo por um conjunto de pessoas, que por algum motivo específico, avaliam com nota baixa, em forma de protesto ou retaliação, algo muito comum na industria cinematográfica. Ambos os filmes possuem, praticamente o mesmo desvio padrão.
