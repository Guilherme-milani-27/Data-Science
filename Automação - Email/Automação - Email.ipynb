# Importando biblioteca

```python
import yfinance
```

# Buscando as cotações de uma ação

```python
ticker = input('Digite o código da ação: ')
dados = yfinance.Ticker(ticker)
```

```python
tabela = dados.history('6mo')
```

# Gerando gráfico

```python
fechamento = tabela.Close
```

```python
fechamento.plot()
```

# Gerando as estatísticas

* Cotação máxima
* Cotação mínima
* Cotação atual

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]
```

# Enviando email

* Abrir uma nova aba no navegador;
* Digitar o endereço do Gmail e digitar enter;
* Clicar em "Escrever";
* Digitar o endereço de email do destinatário;
* Mudar para o campo assunto;
* Digitar o assunto do email;
* Mudar para o campo principal do email;
* Escrever a mensagem do email;
* Clicar em enviar.

```python
import pyautogui
import pyperclip
```

```python
site = 'www.gmail.com'
email = 'milaniguilherme@hotmail.com'
assunto = 'Análise diária'
mensagem = f"""
Bom dia!

Segue abaixo a análise da ação {ticker} dos ultimos seis meses:

Cotação máxima: R${maxima:.2f}
Cotação mínima: R${minima:.2f}
Cotação atual: R${atual:.2f}

Atenciosamente,

Guilherme Milani Silva
"""
```

```python
# Pausa entre as ações
pyautogui.PAUSE = 4

# Abrir uma nova aba no navegador
pyautogui.hotkey('ctrl', 't')

# Copia e cola o endereço do Gmail e digitar "enter"
pyperclip.copy(site)
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('enter')

# Clicar em escrever
pyautogui.click(x=125, y=203)

# Escrever o email do destinátario
pyperclip.copy(email)
pyautogui.hotkey('ctrl', 'v')

# Mudar para o campo assunto
pyautogui.hotkey('tab')

# Escrevendo o assunto do email
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')

# Mudar para o campo principal do email
pyautogui.hotkey('tab')

# Escrever a mensagem do email
pyperclip.copy(mensagem)
pyautogui.hotkey('ctrl', 'v')

# Clicar em enviar
pyautogui.click(x=829, y=697)
```
