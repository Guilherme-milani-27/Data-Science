# Importando a biblioteca Selenium

```python
from selenium import webdriver
```

```python
nav = webdriver.Chrome()
```

# Importando a planilha excel

```python
import pandas as pd
```

```python
tabela = pd.read_excel('commodities.xlsx')
tabela
```

# Buscando cotações na Web

```python
for linha in tabela.index:
    produto = tabela.loc[linha, 'Produto']
    produto = produto.replace('ã', 'a').replace('á', 'a').replace('ó', 'o').replace('ú', 'u').replace('é', 'e').replace('ç', 'c')
    link = f"https://www.melhorcambio.com/{produto.lower()}-hoje"
    nav.get(link)
    preco = nav.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')
    preco = preco.replace('.', '').replace(',', '.')
    tabela.loc[linha, 'Preço Atual'] = float(preco)
display(tabela)
```

```python
tabela['Comprar'] = tabela['Preço Atual'] < tabela['Preço Ideal']
dicionario = {True: 'Comprar', False: 'Não comprar'}
tabela.Comprar = tabela.Comprar.map(dicionario)
tabela
```

# Exportando para excel

```python
tabela.to_excel('commodities_atualizado.xlsx', index=False)
```
